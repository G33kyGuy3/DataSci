```
#Import the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn as skl
import os
```

```
#Load the data
my_data = pd.read_excel("/Employee Attrition Data.xlsx")
```

```
#Store data into a data frame
df1 = pd.read_excel("/Employee Attrition Data.xlsx")
```

```
df1.head(7)
```

```
#Get shape of data
print(df1.shape)
```

```
#Remove some useless columns
df1 = df1.drop('Over18', axis= 1)
df1 = df1.drop('EmployeeNumber', axis= 1)
df1 = df1.drop('StandardHours', axis = 1)
df1 = df1.drop('EmployeeCount', axis = 1)
```

```

#Transform the data
#Transform non-numerical into numerical col
from sklearn.preprocessing import LabelEncoder

for column in df1.columns:
  if df1[column].dtype == np.number:
    continue
  df1[column] = LabelEncoder().fit_transform(df1[column])
```

```
#Create a new column
df1['Age_Year'] = df1['Age']
```

```
#Drop the age column
df1 = df1.drop('Age',axis=1)
```

```
df1
```

```
#Split the data
X = df1.iloc[:,1:df1.shape[1]].values
Y = df1.iloc[:, 0].values
```

```
#Split the data into 75% training and 25% testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.25, random_state = 0)
```

```
from sklearn import svm
#Create a classifer
cls = svm.SVC(kernel="linear")
#train the model
cls.fit(X_train, Y_train)
#predict the response
pred = cls.predict(X_test)
```

```
from sklearn import metrics
#Accuracy
print("Accuracy:", metrics.accuracy_score(Y_test, y_pred=pred))

#Precision Score
print("Precision:", metrics.precision_score(Y_test,y_pred=pred))

#Recall score
print("Recall:",metrics.recall_score(Y_test,y_pred=pred))

print(metrics.classification_report(Y_test,y_pred=pred))
```
